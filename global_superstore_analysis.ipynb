### PHASE 2: Simulate Data Issues (Mess it up intentionally!)

```python

# Introduce nulls in 'Postal Code' and 'Profit'


df.loc[5:50, 'Postal Code'] = np.nan
df.loc[99:143, 'Profit'] = np.nan
df.loc[14442:15000,'Quantity'] = np.nan
```

🧠 np.nan is a special NumPy value used to represent missing data in pandas.

🧠 df.loc[5:50, 'Postal Code'] selects rows 5 to 50 in the 'Postal Code' column.

 2. Add duplicate rows

```python
df = pd.concat([df,df.iloc[[0,10]]],ignore_index=True)
#🧠 df.iloc[[0, 1]] selects the first and second rows.
#🧠 pd.concat([...]) combines original df with the duplicated rows.
```

🔧 3. Create inconsistent text formatting

```python
df.loc[0,'Category'] = 'FURNITURE' # All uppercase
df.loc[1,'Category'] = 'furniture' # All lowecase
```

🧠 This simulates inconsistent capitalization that might affect grouping or filtering operations.

🔧 4. Intentionally corrupt some dates



```python
df.loc[2, 'Order Date'] = 'Not a date'
df['Order Date'] = pd.to_datetime(df['Order Date'], errors= 'coerce')
```

### PHASE 3: Clean the Mess

🧠 df.isna() returns a DataFrame of True/False values where NaN is present.

🧠 .sum() adds up the True values column-wise (since True = 1 in Python).



```python
df.isna().sum()
```

🧠 dropna(subset=['Postal Code']) drops only the rows where 'Postal Code' is NaN.

🔐 We use subset to avoid dropping rows just because other columns are missing.



```python
df = df.dropna(subset=['Postal Code'])
```

🧠 .fillna(0) replaces NaN with 0.

✔️ Business logic matters — in real life, you'd verify with stakeholders whether 0 is appropriate.

```python
#  Fix missing values in Profit

df['Profit'] = df['Profit'].fillna(0)
```

🧠 .drop_duplicates() removes any repeated rows.

```python
df = df.drop_duplicates()
```

🧠 .str.title() capitalizes the first letter of each word — so FURNITURE, furniture → Furniture.

🧠 Useful alternatives:

.str.lower() – all lowercase

.str.upper() – all uppercase



```python
# Standardize Text in Category

df['Category'] = df['Category'].str.title()
df['Category'].unique() #for ensuring operation accuracy
```

🧠 Dates are crucial for time-based analysis — we don’t want invalid ones lingering.

```python
# Fix Broken Dates

df['Order Date'].isna().sum() #confirm how many bad dates were coerced

# dropping bad dates

df = df.dropna(subset=['Order Date'])
```

Univariate Numerical Analysis

##### Sales – Histogram & KDE

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'],bins=110,kde=True,color='blue')
plt.title("Sales Distribution")
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```


Sales – Boxplot (Outlier Detection)



```python
plt.figure(figsize=(6,3))
sns.boxplot(x=df['Sales'],color='lightcoral')
plt.xlim(0,2000) ## Limiting x-axis to make the boxplot more readable
plt.title('Sales Outlier Detection')
plt.xlabel('Sales')
plt.show()
```

 Here we leverage log transformation as you could see in the above chart , data is more concentrated on one side as well as has extreme outliers too

```python
# Avoid errors due to log(0)
df['Log_Sales'] = np.log1p(df['Sales'])  # log1p = log(1 + x)

plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Log_Sales'], color='salmon')
plt.title('Sales Distribution (Log Scale)')
plt.xlabel('Log of Sales')
plt.show()
```

EDA on Profit

 Distribution of Profit

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Profit'],bins=50,kde=True,color='teal')
plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Performed log transformation once again as data was squished and has extreme values on either side

```python
df['Log_Profit'] = np.log1p(df['Profit'])

plt.figure(figsize=(8,4))
sns.boxplot(x=df['Log_Profit'],color='orange')
plt.title('Boxplot of Profit')
plt.xlabel('Profit')
plt.grid(True)
plt.show()
```

Grouped Barplot by Category and Sub-Category

```python
plt.figure(figsize=(14, 6))

# Step 1: Compute total profit per sub-category
profit_by_subcat = df.groupby(['Sub-Category', 'Category'])['Profit'].sum().reset_index()

# Step 2: Sort sub-categories by profit
sorted_subcats = profit_by_subcat.sort_values('Profit', ascending=False)['Sub-Category']

# Step 3: Create the barplot
sns.barplot(
    data=profit_by_subcat,
    x='Sub-Category',
    y='Profit',
    hue='Category',
    order=sorted_subcats,
    palette='Set2',
    errorbar=None
)

# Step 4: Final touches
plt.title('Total Profit by Sub-Category (Sorted) with Category Hue')
plt.xlabel('Sub-Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

 Profit vs Sales Scatter Plot

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='Sales',y='Profit',hue='Category',palette='Set2')
plt.title('Profit vs Sales by Category')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

🧠 Interpretation of the Profit vs Sales Scatter Plot
Insight	What it Means

🔸 Dense Cluster (Low Sales & Low Profit)	Most transactions happen at small scale — low-ticket items sold frequently. These products are the core volume drivers.

🔸 Sparse Upper Region	High-profit, high-sales items are less common — maybe expensive products, sold less often, but very profitable when sold.

🔸 Some Points with High Sales but Negative Profit	Indicates losses despite high sales — possibly due to discounts, high cost, or operational inefficiencies. These products need a closer look.

🔸 Category Color Clusters	If a color (category) dominates the profit zone, it's a strong performer. If it dominates loss zone, it's a red flag.

#### After using log transformation to make the points look more visible 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x = 'Sales', y= 'Profit', hue='Category', alpha = 0.7)

plt.xscale('log')
plt.yscale('log')

plt.title('Profit vs Sales (Log Scale)')
plt.xlabel('Sales (Log Scale)')
plt.ylabel('Profit (Log Scale)')
plt.legend(title='Category', bbox_to_anchor = (1.05,1),loc= 'upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Creating Log Columns of 'Profit' and 'Sales' for avoiding skewness and outliers when visualizing data

```python
# Log-transform Sales and Profit

df['Sales_log'] = np.log1p(df['Sales'])
df['Profit_log'] = np.log1p(df['Profit'])


# Preview the new columns

df[['Sales','Sales_log','Profit','Profit_log']].head()
```

We can now use these new columns in charts to improve clarity and see patterns more clearly.
